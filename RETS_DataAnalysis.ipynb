```python
dataFig5_PAP1corr = pd.io.excel.read_excel("~/PythonNotebooks/RENDR figures/Final Figs/Pub/5/250810_PAP1-qPCR_Colv1805-PAP1RENDR_0vs2-4-5_Corr.xlsx") #path to 250810_PAP1-qPCR_Colv1805-PAP1RENDR_0vs2-4-5_Corr.xlsx


plt.figure(figsize=(4, 4))

plot = sns.regplot(data=dataFig5_PAP1corr, x="FC (PAP1)", y="FC (RETSmScar)")
sns.scatterplot(data=dataFig5_PAP1corr, x="FC (PAP1)", y="FC (RETSmScar)", hue="Sucrose %", palette= "Reds", s=50)
slope, intercept, r, plot, sterr = scipy.stats.linregress(x=plot.get_lines()[0].get_xdata(), y=plot.get_lines()[0].get_ydata())
plt.tight_layout()

x=dataFig5_PAP1corr["FC (PAP1)"]
y=dataFig5_PAP1corr["FC (RETSmScar)"]

x = [i for i in x if not math.isnan(i)]
y = [i for i in y if not math.isnan(i)]

y_pred = []

for i in x:
    y_pred.append(slope*(i) + intercept)

    
plt.annotate("r-squared = {:.3f}".format(r2_score(y, y_pred)), (0, 7.3))
plt.annotate(f"y = {slope:.3f} x + {intercept:.3f}", (0, 8))


plt.show()

```
